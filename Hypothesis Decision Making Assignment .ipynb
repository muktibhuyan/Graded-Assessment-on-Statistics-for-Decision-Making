# Question 1:

```python
# 1. For the suburb Altona, it is postulated that a typical property sells for $800,000.
#Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? 
#Use a significance level of 5%.
```

```python
sub_Altona = df[df['Suburb'] == 'Altona']
sub_Altona
```

```python
mean_price_altona = sub_Altona['Price'].mean()
mean_price_altona
```

```python
# Ho = There is no difference between property price for properties in Altona and $800,000. Ho =800000
# Ha = There is a difference between property price for properties in Altona. 

null_hypo_value = 800000 

t_stat, p_value = stats.ttest_1samp(sub_Altona['Price'], null_hypo_value)

print("T-Statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Property price is not different from $800,000.")
else:
    print("Property price has increased from $800,000.")      

# Since P-value is 30% so we fail to reject the null hypothesis.
```

# Question 2:

```python
# For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? 
#Consider months from October till March as winter months and rest as summer months. 
#Use a significance level of 5%. 
```

```python
df['Date'] = pd.to_datetime(df['Date'])
property_2016 = df[df['Date'].dt.year == 2016]
```

```python
summer = (property_2016['Date'].dt.month >= 4) & (property_2016['Date'].dt.month <= 9)
winter = ~summer
```

```python
# Segregating properties for Summer & winter

summer_properties = property_2016[summer]
winter_properties = property_2016[winter]
```

```python
#Ho = Average property price in summer is equal to the average property price in winter.
# Ha = There is a significant difference in property prices between summer and winter months.

t_statistic, p_value = stats.ttest_ind(summer_properties['Price'], winter_properties['Price'])
print("T-Statistic:", t_statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in property prices between summer and winter months.")
else:
    print("There is no significant difference in property prices between summer and winter months.")
    
# since p-value near to Zero,so we fail to reject the null hypothesis.
```

# Question 3:

```python
#For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? 
#Use the column car in the dataset. Round off your answer to 3 decimal places. 
```

```python
# Property Filter
abbotsford_property = df[df['Suburb'] == 'Abbotsford']
```

```python
total_properties = len(abbotsford_property)
properties_without_parking = len(abbotsford_property[abbotsford_property['Car'] == 0])
proportion_without_parking = properties_without_parking / total_properties
proportion_without_parking
```

```python
from scipy.stats import binom
```

```python
n = 10  # Total number of properties
k = 3   # Number of properties without car parking
p = proportion_without_parking
probability_3_out_of_10 = binom.pmf(k, n, p)*100
rounded_percent = round(probability_3_out_of_10, 3)
print("Probability of exactly 3 out of 10 properties not having car parking:", rounded_percent)
```
